```python
text = open("audio.txt", "r", encoding="UTF-8").read()
text
```

```python
import nltk

# Texto a analizar


# Tokenizar el texto
words = nltk.word_tokenize(text)

# Contar las palabras
word_count = nltk.FreqDist(words)

# Imprimir el total de palabras
print("Total de palabras: ", len(words))

print(sorted(word_count.items(), key=lambda x: x[1], reverse=True))
# Imprimir las palabras y su cantidad
for key,val in word_count.items():
    print(str(key) + ':' + str(val))
```
